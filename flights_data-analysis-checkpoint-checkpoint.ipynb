---------------------------------------------data visualization-------------------------------------------------------------------

```python
#airlines and the  number flight counts.

data['airline'].value_counts(ascending=True).plot.barh(color=['#ff9999','lightblue'])
plt.ylabel("airlines names")
plt.xlabel("number of flights")
plt.title("flights and their counts")
plt.show()
```

```python
data.head()
```

```python
#showing number of the departure and the arrival times of the flights.
departure_time = data['departure_time'].value_counts()
print(departure_time)
```

```python
arrival_time= data['arrival_time'].value_counts()
print(arrival_time)
```

```python
#showing the departure time & arrival time for the flight with their counts.
plt.figure(figsize = (16,4))

plt.subplot(1,2,1)
plt.bar(x=data['departure_time'].value_counts().index,
        height=data['departure_time'].value_counts().values,color=['green','yellow'])
plt.title("DEPARTUR")
plt.xlabel("DEPARTURING-TIME")
plt.ylabel("DEPARTURING-COUNTS")

plt.subplot(1,2,2)
plt.bar(x=data['arrival_time'].value_counts().index,
        height=data['arrival_time'].value_counts().values,color=['brown','pink'])
plt.title("ARRIVAL ")
plt.xlabel("ARRIVAL-TIME")
plt.ylabel("ARRIVAL-COUNTS")
plt.show()
```

```python
data.head()
```

```python
source_city= data['source_city'].value_counts()
print(source_city)
```

```python
destination_city= data['destination_city'].value_counts()
print(destination_city)
```

```python
#showing the destination city and source city for the flights and their counts.
plt.figure(figsize = (16,4))

plt.subplot(1,2,1)
plt.barh(y=data['source_city'].value_counts().index,
        width=data['source_city'].value_counts().values,color=['black','indigo'])
plt.title("SOURCE CITY AND THEIR NUMEBR OF FLIGHTS")
plt.xlabel("NUMBER OF FLIGHTS")
plt.ylabel("SOURCE CITYS")

plt.subplot(1,2,2)
plt.barh(y=data['destination_city'].value_counts().index,
        width=data['destination_city'].value_counts().values,color=['gray','indigo'])
plt.title("ARRIVAL CITY AND THEIR NUMBER OF FLIGHTS")
plt.xlabel("NUMBER OF FLIGHTS")
plt.ylabel("ARRIVAL CITYS")
plt.show()
```

```python
#grouping the airlines and checking their average prices.

data.groupby('airline')['price'].mean()
```

```python
#drawing a categorical plot shows that average ticket price for each airlines and their class.

sns.catplot( x = 'airline' , y = 'price' , kind ='bar' , palette = 'rocket' , data = data,hue = 'class')
plt.show()
```
